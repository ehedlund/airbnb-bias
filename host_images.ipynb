```python
import contextlib
import os.path
import urllib
import pandas as pd
from bs4 import BeautifulSoup
from urllib2 import urlopen, HTTPError, Request
```

```python
def make_soup(url):
    request = Request(url, headers={"User-Agent":'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/66.0.3359.139 Safari/537.36'})

    with contextlib.closing(urlopen(request)) as x:
        html = x.read()

    # html = urlopen(request).read()
    return BeautifulSoup(html, "html5lib")
```

```python
# modified from https://stackoverflow.com/a/23159190
def get_profile_image(user_id, url):
    suffix = "aki_policy=profile_x_medium"

    soup = make_soup(url)
    images = [img for img in soup.findAll('img')]
    image_links = [each.get('src') for each in images]
    for each in image_links:
        if (each.endswith(suffix)):
            filename = "host_images/" + str(user_id) + "_pic.png"
            urllib.urlretrieve(each, filename)
            break
```

```python
def get_listing_ids():
    # get listing ids
    df = pd.read_csv('reviews.csv')
    ids = df['listing_id'].tolist()
    
    # remove duplicates while maintaining order
    unique_ids = []
    for i in ids:
        if i not in unique_ids:
            unique_ids.append(i)
    
    return unique_ids
```

```python
listing_ids = get_listing_ids()
```
